```python
import plotly.graph_objects as go
import numpy as np
filepath = './data/Perovskite_KaiZhu_3DTopography/200112Topography017.txt'
xs,ys,signals = XEItxt_to_arrays(filepath,scales = [1.,1.,0.001], offsets=[0.,0.,0.])

xaxis = sorted(list(set(xs)))
yaxis = sorted(list(set(ys)))
xx,yy = np.meshgrid(xaxis,yaxis)
zz = signals.reshape((len(xaxis),len(yaxis)))

shift_x = 20
shift_y = 8
xx = xx[:50,:50]
yy = yy[:50,:50]
zz = zz[shift_y:50+shift_y,shift_x:50+shift_x]

xx[:,0] = xx[:,1]#-1e-5
xx[:,-1] = xx[:,-2]#+1e-5
yy[0,:] = yy[1,:]#-1e-5
yy[-1,:] = yy[-2,:]#+1e-5

zzmin = np.min(zz)
zz[:,0]=zzmin-0.06
zz[:,-1]=zzmin-0.06
zz[0,:]=zzmin-0.06
zz[-1,:]=zzmin-0.06
zz -= np.min(zz)


# Read data from a csv
#z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')
layout = go.Layout(
             scene=dict(
                 aspectmode='data'
         ))

fig = go.Figure(data=[go.Surface(z=zz,x=xx,y=yy)],layout=layout)

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
class box:
    def __init__(self,length,width,height,corner_pos = (0,0,0),rgb_color=(128,128,128),opacity=1.):
        color_str = "rgb("+str(rgb_color[0])+','+str(rgb_color[1])+','+str(rgb_color[2])+")"
        corner_x,corner_y,corner_z = corner_pos
        xx,yy = np.meshgrid([0+corner_x,1e-5+corner_x,length-1e-5+corner_x,length+corner_x],\
                            [0+corner_y,1e-5+corner_y,width-1e-5+corner_y,width+corner_y])
        cube = np.array([[0,0,0,0],[0,height,height,0],[0,height,height,0],[0,0,0,0]])+corner_z
        self.data = go.Surface(z=cube,x=xx,y=yy,colorscale = [[0,color_str],[1,color_str]],opacity=opacity,lightposition=dict(x=0,y=-1,z=-5))#,surfacecolor=np.ones(cube.shape)+corner_z,cmin=0,cmax=corner_z)
    
```
