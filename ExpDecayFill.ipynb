```python
import numpy as np
import matplotlib.pyplot as plt

A1 = 0.17356
A2 = 0.70699
t1 = 13171.847165
t2 = 600.36966
y0 = 0.0541

t_list = np.arange(0,70000,70000/1400)
y_list = A1*np.exp(-t_list/t1)+A2*np.exp(-t_list/t2)+y0-y0
cutoff = int(5000/50)
plt.plot(t_list[:cutoff],y_list[:cutoff],linewidth=15,color = 'pink')
plt.plot(t_list[cutoff:],y_list[cutoff:],linewidth=15,color = 'skyblue')
# plt.axis('off')
plt.yscale('log')
# plt.ylim(0.02,1.1)
```
